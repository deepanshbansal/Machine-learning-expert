## Standardise the data using Standard Scaler


```python
train_data=strat_train_set.iloc[:,0:13]
train_label=strat_train_set.iloc[:,13]
test_data=strat_test_set.iloc[:,0:13]
test_label=strat_test_set.iloc[:,13]
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
scaled_train_data=sc.fit_transform(train_data)
scaled_train_data
```

## creating model for predictions


```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
#model=LinearRegression()
#model=DecisionTreeRegressor()
model=RandomForestRegressor()
model.fit(scaled_train_data,train_label)
```

```python
testing_data=sc.fit_transform(test_data)
```

```python
pred=model.predict(testing_data)
pred
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
lin_mse=mean_squared_error(pred,test_label)
lin_rmse=np.sqrt(lin_mse)
print(lin_mse)
```

```python
lin_rmse
```

## Using better evaluation technique  Cross Validation

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(model,train_data,train_label,scoring='neg_mean_squared_error',cv=10)
rmse_scores=np.sqrt(-scores)
rmse_scores
```

```python
def print_scores(scores):
    print('Scores :',scores)
    print('Mean :',scores.mean())
    print('Standard Deviation:',scores.std())
```

```python
print_scores(rmse_scores)
```
